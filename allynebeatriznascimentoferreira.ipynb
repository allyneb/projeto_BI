```python
import pandas
import csv
from random import randint
```

```python
# ponto 1

format_data = pandas.read_csv("selecaoEstagio.csv", delimiter=";")
```

```python
# ponto 2 - linhas e colunas

rows = len(format_data)
columns = len(format_data.columns)
print(rows)
print(columns)
```

```python
# ponto 2 - valores unicos

for column in format_data.columns:
    print(format_data[column].unique())
```

```python
# ponto 2 - valores nulos

for i in range(rows):
    print((format_data.loc[i].values == '').sum())
```

```python
# ponto 3

numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numerical_format_data = format_data.select_dtypes(include=numerics)
average = numerical_format_data.mean(axis=0)
median = numerical_format_data.median(axis=0)
std = numerical_format_data.std(axis=0)
print(average)
print(median)
print(std)
print(numerical_format_data)
```

```python
# ponto 4

numerical_format_data = numerical_format_data.drop(numerical_format_data.columns[-2:],axis=1)
print(numerical_format_data)
```

```python
# ponto 5

def random_rows(data: pandas.DataFrame):
    print(data.sample(randint(0, len(data))))
    return data.sample(randint(0, len(data)))

random_rows(format_data)

```
